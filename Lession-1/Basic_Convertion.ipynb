```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("/content/drive/MyDrive/Colab Notebooks/bird.jpg")  #Load Image
```

```python
print("Image Shape",img.shape)   #Check Shape
```

```python
img  # 3D array Value
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)   # convert CV(BGR) to matplotlib(RGB)
```

```python
plt.imshow(img)  # show CV image
```

```python
plt.imshow(img_rgb)   # show matplotlib image
plt.axis("off")
plt.show()
```
