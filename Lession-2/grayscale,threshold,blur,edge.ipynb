```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("/content/drive/MyDrive/Colab Notebooks/children.jpg")  #Load Image
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)   # convert CV(BGR) to matplotlib(RGB)
```

```python
plt.imshow(img)  # show CV image
```

```python
plt.imshow(img_rgb)   # show matplotlib image
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)  #convert in gray scale
```

```python
# simple Thresholding
_,thresh = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)  #convert in binary
```

```python
# Adaptive Thresholding
adaptive = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,1)
```

```python
# Blur
blur_avg=cv2.blur(gray,(5,5))
blur_gauss=cv2.GaussianBlur(gray,(5,5),0)
blur_median=cv2.medianBlur(gray,5)
```

```python
#Edge Detection
edges=cv2.Canny(gray,100,200)
```

```python
#Result
titles=['Original','Gray','Threshold','Adaptive','Blur_Avg','Blur_Gauss','Blur_Median','Edges']
images=[img_rgb,gray,thresh,adaptive,blur_avg,blur_gauss,blur_median,edges]
```

```python
plt.figure(figsize=(30,20))
for i in range(8):
  plt.subplot(2,4,i+1)
  if i==0:
    plt.imshow(images[i])
  else:
    plt.imshow(images[i],cmap='gray')
  plt.title(titles[i])
  plt.axis('off')
plt.show()
```
